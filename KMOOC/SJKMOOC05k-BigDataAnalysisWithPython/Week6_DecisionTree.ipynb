## Decision Tree

- 의사결정 트리 또는 의사결정 나무(Decision Tree)는 기계학습(Machine Learning)에서 지도학습(Supervised Learning)의 알고리즘으로 분류(Classification) 또는 회귀(Regression)분석 목적으로 사용
- 의사결정 규칙(Decision Rule)을 나무구조 표현을 통해 분류와 예측을 수행하는 분석 방법
- 분류 또는 예측 과정이 나무구조로 표현되어 비교적 쉽게 이해
- 목표변수 유형에 따른 의사결정 트리
    - 범주형 목표변수 : 분류 트리(Classification Tree)
        - 목표변수가 이산형인 경우, 각각의 범주에 속하는 빈도에 기초해 분리 발생 → 분류 트리 구성
    - 연속형 목표변수 : 회귀 트리(Regression Tree)
        - 목표변수가 연속형인 경우, 평균과 표준편차에 기초해 분리 발생 → 회귀 트리 구성
        
![week6_1.png](images/week6_1.png)    
- 뿌리 마디(Root Node)
    - 트리 구조가 시작되는 마디, 전체 자료로 구성
- 부모 마디(Parent Node) 
    - 자식 마디의 상위 마디
- 자식 마디(Child Node)
    - 하나의 마디로부터 분리되어 나간 2개 이상의 마디들을 의미
- 끝 마디(Terminal Node) 또는 잎(Leaf Node)
    - 트리 줄기의 끝에 위치하고 있고 자식 마디가 없는 마디
- 가지(Branch)
    - 뿌리 마디로부터 끝 마디까지 연결된 마디들
- 깊이(Depth)
    - 뿌리 마디로부터 끝 마디를 이루는 층의 수
    
#### Decision Tree 분리기준(Split Criterion)
- 부모 마디(노드)로부터 자식 마디(노드)들이 형성될 때, 생성된 자식 노드에 속하는 자료의 순수도(Purity)가 가장 크게 증가하도록 트리를 형성하며 진행
- 입력 변수를 이용해 목표 변수의 분포를 얼마나 잘 구별하는 정도를 파악해 자식 마디가 형성되는데, 목표 변수의 구별 정도를 불순도(Impurity, 다양한 범주들의 개체들이 포함되어 있는 정도)에 의해 측정

#### Decision Tree 분리기준(Split Criterion)
- 지니 지수(Gini Index)
    - 데이터 집합의 불순도를 측정
    - 지니 지수는 0~1 사이의 값을 가지며, 어떤 데이터 집합에 속한 개체(레코드)들이 같은 범주(클래스)로 구성되어 있으면 지니 지수는 최솟값이 0을 갖고 해당 데이터 집합은 순수하다고 볼 수 있음
    - 즉, 지니 지수가 작을수록 잘 분류된 것으로 볼 수 있음
    > $G = 1 - \sum_{j=1}^c P(j)^2 = 1 - \sum_{j=1}^c{\left( {n_j} \over{n} \right)}^2$
        - $C$ : 범주(클래스)의 수
        - $P(j)$ : $j$번째 범주(클래스)에 분류될 확률
        - $n$ : 노드에 속하는 개체 수
        - $n_j$ : 노드에 속하는 수 중 $j$번째 범주(클래스)에 속하는 개체수
    	
- 엔트로피 지수(Entropy Index)
    - 엔트로피는 주어진 데이터 집합의 혼잡도를 의미
    - 주어진 데이터 집합에 서로 다른 범주(클래스)의 개체(레코드)들이 많이 섞여 있으면 엔트로피가 높고, 같은 범주의 개 체들이 많이 있으면 엔트로피가 낮음
    - 엔트로피 지수는 0~1 사이의 값을 가지며, 가장 혼잡도가 높은 상태(서로 다른 범주의 개체들이 섞여 있는 상태)는 1, 혼잡도가 가장 낮은 상태(하나의 범주의 개체로 구성된 상태)는 0
    > $H(S) = -p_1\log_2p1 - p2\log_2p2$
- 정보 이득(Information Gain)
    - 상위 노드의 엔트로피 지수에서 하위 노드의 가중평균한 엔트로피 지수를 뺀 것을 의미
    - 즉, 원래 상위 노드의 엔트로피를 구하고 어떤 속성을 선택한 후의 x개의 하위 노드로 분리된 것에 대한 가중평균한 엔트로피를 구한 값의 차를 의미(계산되어 나온 값이 클 수록 정보 이득이 큰 것을 의미, 선택한 어떤 속성이 분류하기 좋다고 볼 수 있음)
    > $\text{Gain}(S, A) = H(S) - \sum_v {\left\vert S_v \right\vert \over \left\vert S \right\vert} \text{Entropy}(S_v)$

#### 예제를 이용한 지니계수 계산
![week6_2.png](images/week6_2.png)    
Outlook 속성으로 분리하는 경우(Left = {Sunny}, Right = {Overcast, Rain} 일때)
- Left(Sunny)의 지니계수
> $1 - \left\{ \left( {3 \over 5} \right)^2 + \left( {2 \over 5} \right)^2 \right\} = 0.48$
- Right (Overcast, Rain)의 지니 지수 계산
> $1 - \left\{ \left( {4 \over 9} \right)^2 + \left( {5 \over 9} \right)^2 \right\} = 0.4928$
- 지니지수는 작을수록 노드 구성이 잘된 것을 의미
- Left 노드의 지니 지수는 0.48, Right 노드의 지니 지수 는 0.4928로 Right 노드보다 지니 지수가 더 작은 Left 노드가 더 구성이 잘된 것을 알 수 있다.
- 즉, 왼쪽 노드에 sunny를 배치시키고, 오른쪽 노드에는 overcast와 rain을 배치한다.

```python
from sklearn.metrics import classification_report, confusion_matrix #sklearn.metrics: 모형평가에 사용되는 서브 패키지
#classification_report : 주요 분류 측정 항목을 보여주는 보고서 모듈 
#confusion_matrix : 분류의 정확성을 평가하기 위한 오차행렬 계산 모듈
from sklearn.model_selection import train_test_split #sklearn.model_selection : 클래스를 나눌 때, 함수를 통해 train/test를 나눌 때, 모델 검증에 사용되는 서브 패키지
#train_test_split : 배열 또는 행렬을 임의의 훈련(train) 및 테스트(test) 하위 집합으로 분할하는 모듈
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

from IPython.display import Image #IPython.display : IPython 내에 정보를 보여주는 도구용 공용 API 
#Image : raw 데이터가 있는 PNG, JPEG 이미지 객체를 만드는 모듈

import pandas as pd
import numpy as np
import pydotplus #pydotplus : 그래프를 생성하는 graphviz의 Dot 언어를 파이썬 인터페이스에 제공하는 모듈
import os #os : 운영체제(Operating System)와 상호작용하기 위한 기본적인 기능(경로 생성, 변경 등)이 제공되는 모듈
```

```python
tennis_data = pd.read_csv("playtennis.csv")
tennis_data.head()
```

```python
##전처리 : 문 자열(String) 타입에서 숫자(int) 타입으로 대치(replace)해 변수 tennis_data의 각 컬럼별 저장

tennis_data.Outlook = tennis_data.Outlook.replace("Sunny", 0)
tennis_data.Outlook = tennis_data.Outlook.replace("Overcast", 1)
tennis_data.Outlook = tennis_data.Outlook.replace("Rain", 2)

tennis_data.Temperature = tennis_data.Temperature.replace("Hot", 3)
tennis_data.Temperature = tennis_data.Temperature.replace("Mild", 4)
tennis_data.Temperature = tennis_data.Temperature.replace("Cool", 5)

tennis_data.Humidity = tennis_data.Humidity.replace("High", 6)
tennis_data.Humidity = tennis_data.Humidity.replace("Normal", 7)

tennis_data.Wind = tennis_data.Wind.replace("Weak", 8)
tennis_data.Wind = tennis_data.Wind.replace("Strong", 9)

tennis_data.Class = tennis_data.Class.replace("Yes", 10)
tennis_data.Class = tennis_data.Class.replace("No", 11)

tennis_data.head()
```

```python
X = np.array(pd.DataFrame(tennis_data, columns=["Outlook", "Temperature", "Humidity", "Wind"]))
#tennis_data의 해당 컬럼의 값들을 데이터프레임 형태로 추출해 np.array로 저장
y = np.array(pd.DataFrame(tennis_data, columns=["Class"]))
#tennis_data의 해당 컬럼의 값을 데이터프레임 형태로 추출해 np.array로 저장

#컬럼 지정하지 않고 np.array(pd.DataFrame(tennis_data))로 하면 모든 값이 저장 된다.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y) #각 데이터를 훈련과 테스트용으로 구분해 저장
#일반적으로 train / test의 비율 = train(7.5) : test(2.5)
```

```python
dt_clf = DecisionTreeClassifier() #결정트리 모델 생성
dt_clf = dt_clf.fit(X_train, y_train) #학습
dt_prediction = dt_clf.predict(X_test) #테스트 데이터로 값 예측

print(confusion_matrix(y_test, dt_prediction)) #실제값과 예측값으로 오차 행렬 출력
```

```python
print(classification_report(y_test, dt_prediction))
```

```python
feature_names = tennis_data.columns.tolist()
#트리 표현 함수(tree.export_graphviz())에 입력되는 파라미터 입력 위해
feature_names = feature_names[0:4] #Outlook, Temperature, Humidity, Wind만 추출
target_names = np.array(["Play No", "Play Yes"]) #트리 표현 함수(tree.export_graphviz())에 입력되는 파라미터 입력 위해

#의사결정 트리를 dot형식으로 내보내는 함수인 export_graphviz()를 이용
dt_dot_data = tree.export_graphviz(dt_clf, out_file=None, #파일이나 문자열 변환. None일 경우 문자열로
                                   feature_names=feature_names, #features이름
                                   class_names=target_names, #class 이름
                                   filled = True, #True일 경우 분류 위한 색칠 여부
                                   rounded = True, #둥근 모서리
                                   special_characters = True) #True일 경우 특수 문자 표시
dt_graph = pydotplus.graph_from_dot_data(dt_dot_data) #그래프 로드
#Pydotplus 모듈 중 Dot 형식의 데이터로 정의된 그래프를 로드(load)하는 함수인 graph_from_dot_data()에 변수 dt_dot_data를 입력한 후 변수 dt_graph에 저장
```

## Random Forest
----

- Random Forest는 2001년에 Leo Breiman에 의해 처음으로 소개
- Decision Tree의 단점을 개선하기 위한 알고리즘 중 하나
- Decision Tree의 확장

![week6_3.png](images/week6_3.png)    
1. Dataset에서 샘플 데이터를 선택(bagging/bootstrap aggregation)
2. 샘플 데이터를 이용해 Decision Tree를 생성
3. 과정 1), 2)를 n번 반복
4. 과정 3)을 통해 생성한 n개의 Decision Tree 를 이용해 예측 (Tree를 모아 Forest를 만든다.)
5. 예측 결과에서 가장 많이 등장하는 결과를 선택하여 최종 결과로 선택

#### Random Forest의 특징
- 여러 개의 Decision Tree를 결합함으로 단일 Decision Tree의 결점을 극복 
- Over-fitting 문제가 적음
- 구현이 간단함
- 병렬 계산이 간편함

#### Random Forest에서의 2가지 random
- Dataset에서 샘플 데이터를 random으로 선택
    ![week6_4.png](images/week6_4.png)    
    - Random으로 n개의 데이터 선택
    - 선택한 n개의 데이터는 중복이 가능
    - 선택한 t개의 샘플데이터 사이 데이터 중복가능 
    - 선택한 샘플 데이터로 Decision Tree 생성(특징의 선택 및 특징의 순서를 무작위로 진행됨)
- 샘플 데이터에서 feature(특정값)을 random으로 선택해 decision tree를 생성
    ![week6_5.png](images/week6_5.png)    
    - 선택하는 feature의 개수는 Sqrt(전체 feature 수), Log2(전체 feature 수)등 방법으로 계산
    
#### 자기성능평가
- Bagging
    - 63%의 데이터를 이용해 매개 tree를 생성
    - 나머지 37%의 데이터를 이용해 매개 tree의 성능을 평가
    - 매개 tree에 입력하는 데이터는 다름
- Out-of-Bag(OOB)
    - OOB 데이터를 이용해 tree의 성능을 교정
    - OBB는 성능 통계에서 많이 사용됨    

```python
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score #성능 평가 모듈
from sklearn.ensemble import RandomForestClassifier

import numpy as np
import pandas as pd
```

```python
iris = load_iris() #데이터 불러오기 #전체 150개

x_train = iris.data[:-30]
y_train = iris.target[:-30]
x_test = iris.data[-30:] 
y_test = iris.target[-30:]

print(y_train)
print(y_test)

#데이터의 분리가 잘 되지 않음
```

```python
rfc = RandomForestClassifier(n_estimators=10) #10개의 tree가진 random forest 생성
#n_estimators : Decision Tree의 개수 (default=10)
#max_features : 최대 고려하는 feature의 개수(default=auto) 
#oob_score : out-of-bag(OOB) 사용 여부(default=False)
#http://scikit- learn.org/stable/modules/generated /sklearn.ensemble.RandomForestClassifier.html
rfc
```

```python
rfc.fit(x_train, y_train) #모델 학습
prediction = rfc.predict(x_test) #테스트. 예측
print(prediction==y_test) #30중 8개 예측 실패
```

```python
rfc.score(x_test, y_test) #테스트 데이터에 대한 예측 결과의 정확도
```

```python
#sklearn.metrics : scikit-learn 패키지 중 모듈 성능 평가방법 모듈
 
print("Accuracy is :", accuracy_score(prediction, y_test)) #예측, 실제 값
print(classification_report(prediction, y_test)) #예측, 실제값

## 위에서 데이터의 분리를 제대로 하지 않았기 때문에
```

```python
##성능재고를 위한 데이터 재 분리

x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2) #데이터 무작위 혼합 후 train, test를 8:2로
print(y_train)
print(y_test)

#분리가 잘 되었다.
```

```python
clf = RandomForestClassifier(n_estimators=10)
clf.fit(x_train, y_train)
prediction_1 = clf.predict(x_test)

print("Accuracy is :", accuracy_score(prediction_1, y_test)) #예측, 실제 값
print(classification_report(prediction_1, y_test)) #예측, 실제값

## 데이터의 분리를 제대로 한 후, 정확도가 대폭 개선되었다.
```

```python
clf_2 = RandomForestClassifier(n_estimators=200, max_features=4, oob_score=True)
clf_2.fit(x_train, y_train)
prediction_2 = clf_2.predict(x_test)

print("Accuracy is :", accuracy_score(prediction_2, y_test))
print(classification_report(prediction_2, y_test))
#Data set의 size가 작으면 성능 차이가 크게 없다. 
```

```python
## 각 feature의 중요도 확인
for feature, imp in zip(iris.feature_names, clf_2.feature_importances_):
    print(feature, imp)
#RandomForestClassifier()의 파라미터 ‘oob_score=True’ 선택해야 알 수 있다.
```

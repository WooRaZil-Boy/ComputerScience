```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline 
#ipython의 magic function 중 하나로 matplotlib의 시각화 결과를 ipython notebook 안에서 출력
```

```python
value = np.random.standard_normal(40) #표준 정규 분포 (평균 = 0, 표준 = 1)에서 40개 난수 가져오기

value #np.ndarray로 생성된다.
```

```python
index = range(len(value))
plt.plot(index, value) #x, y축에서 나타낸 변수들

#x/y 축 범위 설정
plt.xlim(0, 20) #x축 범위 설정
plt.ylim(np.min(value)-1, np.max(value)+1) #y축 범위 설정

#plt.figure() : plot의 새로운 모양을 생성하는 함수
plt.figure(figsize=(7, 4)) #튜플로 그래프 사이즈 결정
plt.plot(value.cumsum(), "b", lw=1.5) #하나의 그래프에서 각 각 plot을 그린다. cumsum()은 누적합. b는 색. lw는 스타일 lineWidth
plt.plot(value.cumsum(), "ro") #하나의 그래프에서 각 각 plot을 그린다.
plt.xlabel("index")
plt.ylabel("value")
plt.title("Line Plot 1")
```

```python
#각 patch에 모두 다른 색

value = np.random.standard_normal(500) #정규분포에서 500개 난수
cm = plt.cm.get_cmap("Spectral") #plt.cm: 다양한 colormap들과 관련 함수를 제공
plt.figure(figsize=(12, 6))

#Plot histogram
n, bins, patches = plt.hist(value, bins=30, color="green") #히스토그램이 반환하는 값들을 각각 n, bins, pathches에 저장
print("1. bins: " + str(bins)) #bins: bin의 엣지(모서리). 총 31개의 edge가 존재(bins가 30이므로)
print("2. the length of bins: " +  str(len(bins)))

bin_centers = 0.5*(bins[:-1]+bins[1:]) #중간값 구하기. bins[:-1] = bins배열의 첫 번째 ~ 마지막 요소 전. bins[1:] = bins배열의 두 번째 ~ 마지막 요소
print("3. bin_centers: " + str(bin_centers))

#정규화 : 데이터의 전체 구간을 정해진 범위 안으로 변환하여 데이터를 관찰하는 방법
col = (bin_centers - min(bin_centers))/(max(bin_centers) - min(bin_centers))
#최대-최소 정규화 = (현재값 - 최소값) / (최대값 - 최소값)
print("4. col: " + str(col))

for c, p in zip(col, patches): #zip으로 묶인 두 개의 변수에 반복자 생성
    plt.setp(p, "facecolor", cm(c)) #변수의 개수만큼 setp 호출

plt.xlabel("value")
plt.ylabel("frequency")
plt.title("Histogram 3")
```

```python
import seaborn as sns

data2 = sns.load_dataset("flights") #seaborn 내장 데이터 셋 #pd.DataFrame으로 반환

data2.head()
```

```python
#위의 데이터프레임을 히트맵으로 분석하기 위해 연도별 데이터로 정리해야 한다.(index를 year, column을 month, value를 passengers로 바꿔야 한다.)
pivoted_data = data2.pivot("year", "month", "passengers") #pandas.DataFrame.pivot(index, column, value)데이터 모양 변경
#https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.pivot.html

pivoted_data
```

```python
sns.set(context="poster", font="monospace") #시각화 옵션
sns.heatmap(pivoted_data, annot=True, fmt="d", linewidths=3) #데이터, annot:각 셀의 데이터 표시 여부, 값의 형태(d: 10진수)
```

```python
#옵션 초기화하기
sns.set() #빈 파라미터
sns.reset_orig() #reset prams to defaults
sns.heatmap(pivoted_data, annot=True, fmt="d", linewidths=3) #데이터, annot:각 셀의 데이터 표시 여부, 값의 형태(d: 10진수)
```

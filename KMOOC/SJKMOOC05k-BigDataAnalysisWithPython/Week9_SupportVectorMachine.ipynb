#### Soft margin Linear SVM
![week9_12.png](images/week9_12.png)    
- Hard marin(모든 입력 벡터들을 초평면 사이에 두고 무조건 한 그룹에 속하게 분류함: 매우 엄격하게 두 개의 그룹으로 분리하는 초평면을 구함) 방식을 보완
- 서포트 벡터가 위치한 경계선에 Slack Variable(여유 변수)를 둔 방식
- 선형으로 분리할 수 없는 데이터일 때 사용(잡음이 있는 측정값이 있거나 데이터가 비선형)
- 잘못 분류 된 경우 초평면에서 거리를 생각할 수 있는 여유 변수(Slack Variable)를 각 요소 에 할당하고, 그렇지 않은 경우 0을 지정
![week9_13.png](images/week9_13.png)    

- C값이 매우 큰 경우 Soft margin SVM은 Hard margin SVM과 같음
- C값이 매우 작으면 우리는 훈련 데이터에서 잘못된 분류를 볼 수 있음

```python
"""
SVM에서 사용되는 Parameter는 Kernel, C, gamma 대표적으로 많이 사용하는 총 세가지 종류가 있음
- Kernel
    • 커널의 종류 설정
    • 기본값으로 rbf를 설정(값은 특정 중심에서 거리에 의존하는 함수 값)
    • rbf(Radial Basis Function, linear, poly(Polynomial) 총 3가지가 있음
- C
    • 얼마만큼의 여유를 가지고 오류를 인정할 건지에 대한 값
    • 기본값으로 1.0 설정
    • C 값을 낮추면 초평면이 매끄러워지고, 값을 높이면 서포트 벡터들을 더 잘 분류함 (Classifying training points correctly)
- Gamma
    • 기본값으로 auto 설정
    • Gamma 값을 낮추면 초평면에서 멀리 떨어진 서포트 벡터들의 영향이 낮고 값을 높이면 멀리 떨어진 요소들의 값이 영향이 큼
    • 값을 높일 경우 초평면에 인접한 서포트 벡터들의 영향(Weight)가 커지기 때문에 초평면이 울퉁불퉁(Uneven)하게 됨
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #matplotlib의 서브패키지로 *MATLAB 처럼 플롯을 그려주는 패키지
#*MATLAB : 수치 해석 및 프로그래밍 홖경을 제공하는 공학용 소프트웨어
from sklearn import svm, datasets #이렇게 한 번에 여러개 가져올 수도 있다.
%matplotlib inline 
#%matplotlib inline : ipython의 “magic function” 중 하나로, matplotlib의 시각화 결과를 Ipython notebook 안에서 출력하는 함수
```

```python
iris = datasets.load_iris()
print(iris.keys()) #딕셔너리 형식에 데이터에서 key값만 출력
print(iris.data.shape) #data에 대핚 전체 행, 열 길이를 출력 (150개의 자료, 4개의 column)
print(iris.feature_names) #feature(column) name 출력
#'sepal length (cm)' : 꽃받침 길이
#'sepal width (cm)' : 꽃받침 너비
#'petal length (cm)' : 꽃잎 길이
#'petal width (cm)' : 꽃잎 너비
```

```python
print(iris.DESCR) #데이터에 세부적인 내용을 저장하고 있는 DESCR을 출력
#(sklearn dataset은 해당 dataset에 세부정보를 제공)
```

```python
## 데이터 학습 시키기
x = iris.data[:, :2] #전체 행과 2개의 열 #처음 2개의 column만 있는 전체 데이터
y = iris.target #target(정답)
SVM = svm.SVC(kernel="linear", C=1).fit(x, y) #sklearn.svm.SVC() : SVM 모델 생성 #학습
# (1) kernel : 차원의 세분화를 함. 대표적으로 rbf(Radial Basis Function), linear, poly(Polynomial)을 사용. 기본값 rbf
# (2) C : 값을 낮추면 초평면이 매끄러워지고, 값을 높이면 서포트 벡터들을 더 잘 분류함. 기본값 1.0
#         C값을 작게 하면 한쪽의 오분류를 낮추는 대신에 다른쪽 오분류가 높아지고, C값을 크게 하면 양쪽의 오분류를 비슷하게 허용
# (3) gamma : 값을 낮추면 초평면에서 멀리 떨어진 서포트 벡터들의 영향이 낮고 값을 높이면 멀리 떨어진 요소들의 값이 영향이 커짐. 기본값 auto
```

```python
## 데이터 시각화 전처리
x_min, x_max = x[:, 0].min()-1, x[:, 0].max()+1 #첫째 column값 중 최소값 -1, 최대값 +1 #numpy.min(), numpy.max()
y_min, y_max = x[:, 1].min()-1, x[:, 1].max()+1 #둘째 column값 중 최소값 -1, 최대값 +1 #numpy.min(), numpy.max() #타겟이 아님
#plot의 처음과 끝을 지정해 주기 위한 전처리

plot_unit = 0.025 #축의 단위 #x_max 값을 x_min 값으로 나눈 후에 100으로 다시 나누어 나온 값 0.025
#축의 단위는 plot을 그릴 때 배경의 세밀한 정도를 나타내기 때문에 그 수치를 낮게 잡아서 세밀하게 확인을 한다.
#방안지의 눈금 하나의 크기. (가로축/세로축 각가의 좌표 크기 : 0.025)
#x_min과 x_max 사이에 plot_unit(0.025)만큼의 간격을 둔 숫자를 생성
#y_min과 y_max 사이에 plot_unit(0.025)만큼의 간격을 둔 숫자를 생성
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_unit), np.arange(y_min, y_max, plot_unit))
#xx : x가 y의 차원 크기의 행의 개수 만큼 반복된 값을 가지고 있는 변수
#yy : y가 x의 차원 크기의 열의 개수 만큼 반복된 값을 가지고 있는 변수
#np.meshgrid() : 1차원 배열형태의 데이터들이 주어지면 2차원 배열형태로 교체해주는 함수
#np.arange() : 지정된 간격 내에서 균등하게 간격을 둔 값을 돌려줌 #python의 range()와 비슷

#https://datascienceschool.net/view-notebook/17608f897087478bbeac096438c716f6/
```

```python
print("x_min : ", x_min)
print("x_max : ", x_max)
print("y_min : ", y_min)
print("y_max : ", y_max)
print("xx : ", xx) #x_min에서 x_max까지 plot_unit씩 증가 (가로: 가로축이므로) #2차원 #데이터 수 150에서 좌우 여분만큼 조금 더 증가
print("yy : ", yy) #y_min에서 y_max까지 plot_unit씩 증가 (세로: 세로축이므로) #2차원 #데이터 수 150에서 상하 여분만큼 조금 더 증가
```

```python
#ravel() : 두 개의 배열을 합쳐서 하나의 배열로 만드는 기능
#numpy.c_[ , ] : 두 개의 배열 쌍을 합쳐 하나의 배열을 구성

x1 = np.array([[1, 2, 3], [1, 2, 3]])
y1 = np.array([[4, 5, 6], [4, 5, 6]])
print(np.ravel(x1))
print(np.c_[x1.ravel(), y1.ravel()])
```

```python
z = SVM.predict(np.c_[xx.ravel(), yy.ravel()]) #예측 #np.c_[] : 배열에 열을 추가할 때 사용
#학습된 SVM모델에 데이터를 입력해서 입력 데이터가 어떤 class 값인지 분류한 값을 저장. (xx값에 저장된 값과 같은 차원으로 교체한 데이터를 저장)
#Numpy.ravel() : 행렬의 1행부터 순차적으로 값을 불러와서 1차원 배열을 만든다 (평탄화. flatten과 유사)

print(np.c_[xx.ravel(), yy.ravel()]) #xx와 yy를 하나씩 매칭
print("z : ", z)
print("z.shape : ", z.shape)

z = z.reshape(xx.shape) #차원 다시 변경
print()
print("********** AFTER RESHAPE **********")
print()

print("z.shape : ", z.shape)
print("xx.shape : ", xx.shape)
print("z : ", z)
```

```python
plt.pcolormesh(xx, yy, z, alpha=0.1) #Matplotlib.pyplot.pcolormesh() : 기본적으로 3개의 배열을 가지고 heatmap 생성
#xx, yy좌표에 해당하는 z값에 따라 색상을 다르게 적용. 투명도는 0.1
plt.scatter(x[:, 0], x[:, 1], c=y) #x축을 첫 번째 컬럼, y축을 두 번째 컬럼. c=색상(target(y)에 따라 다름)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title("Support Vector Machine")
plt.show()
print("accurary : ", SVM.score(X = x, y = y)) #sklearn.metrics.score : 샘플 데이터를 통해 예측한 결과 값과 정확한 결과 값을 비교해서 성능을 평가
```

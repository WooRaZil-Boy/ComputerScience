```python
"""
SVMì—ì„œ ì‚¬ìš©ë˜ëŠ” ParameterëŠ” Kernel, C, gamma ëŒ€í‘œì ìœ¼ë¡œ ë§ì´ ì‚¬ìš©í•˜ëŠ” ì´ ì„¸ê°€ì§€ ì¢…ë¥˜ê°€ ìˆìŒ
- Kernel
    â€¢ ì»¤ë„ì˜ ì¢…ë¥˜ ì„¤ì •
    â€¢ ê¸°ë³¸ê°’ìœ¼ë¡œ rbfë¥¼ ì„¤ì •(ê°’ì€ íŠ¹ì • ì¤‘ì‹¬ì—ì„œ ê±°ë¦¬ì— ì˜ì¡´í•˜ëŠ” í•¨ìˆ˜ ê°’)
    â€¢ rbf(Radial Basis Function, linear, poly(Polynomial) ì´ 3ê°€ì§€ê°€ ìˆìŒ
- C
    â€¢ ì–¼ë§ˆë§Œí¼ì˜ ì—¬ìœ ë¥¼ ê°€ì§€ê³  ì˜¤ë¥˜ë¥¼ ì¸ì •í•  ê±´ì§€ì— ëŒ€í•œ ê°’
    â€¢ ê¸°ë³¸ê°’ìœ¼ë¡œ 1.0 ì„¤ì •
    â€¢ C ê°’ì„ ë‚®ì¶”ë©´ ì´ˆí‰ë©´ì´ ë§¤ë„ëŸ¬ì›Œì§€ê³ , ê°’ì„ ë†’ì´ë©´ ì„œí¬íŠ¸ ë²¡í„°ë“¤ì„ ë” ì˜ ë¶„ë¥˜í•¨ (Classifying training points correctly)
- Gamma
    â€¢ ê¸°ë³¸ê°’ìœ¼ë¡œ auto ì„¤ì •
    â€¢ Gamma ê°’ì„ ë‚®ì¶”ë©´ ì´ˆí‰ë©´ì—ì„œ ë©€ë¦¬ ë–¨ì–´ì§„ ì„œí¬íŠ¸ ë²¡í„°ë“¤ì˜ ì˜í–¥ì´ ë‚®ê³  ê°’ì„ ë†’ì´ë©´ ë©€ë¦¬ ë–¨ì–´ì§„ ìš”ì†Œë“¤ì˜ ê°’ì´ ì˜í–¥ì´ í¼
    â€¢ ê°’ì„ ë†’ì¼ ê²½ìš° ì´ˆí‰ë©´ì— ì¸ì ‘í•œ ì„œí¬íŠ¸ ë²¡í„°ë“¤ì˜ ì˜í–¥(Weight)ê°€ ì»¤ì§€ê¸° ë•Œë¬¸ì— ì´ˆí‰ë©´ì´ ìš¸í‰ë¶ˆí‰(Uneven)í•˜ê²Œ ë¨
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #matplotlibì˜ ì„œë¸ŒíŒ¨í‚¤ì§€ë¡œ *MATLAB ì²˜ëŸ¼ í”Œë¡¯ì„ ê·¸ë ¤ì£¼ëŠ” íŒ¨í‚¤ì§€
#*MATLAB : ìˆ˜ì¹˜ í•´ì„ ë° í”„ë¡œê·¸ë˜ë° í™–ê²½ì„ ì œê³µí•˜ëŠ” ê³µí•™ìš© ì†Œí”„íŠ¸ì›¨ì–´
from sklearn import svm, datasets #ì´ë ‡ê²Œ í•œ ë²ˆì— ì—¬ëŸ¬ê°œ ê°€ì ¸ì˜¬ ìˆ˜ë„ ìˆë‹¤.
%matplotlib inline 
#%matplotlib inline : ipythonì˜ â€œmagic functionâ€ ì¤‘ í•˜ë‚˜ë¡œ, matplotlibì˜ ì‹œê°í™” ê²°ê³¼ë¥¼ Ipython notebook ì•ˆì—ì„œ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜
```

```python
iris = datasets.load_iris()
print(iris.keys()) #ë”•ì…”ë„ˆë¦¬ í˜•ì‹ì— ë°ì´í„°ì—ì„œ keyê°’ë§Œ ì¶œë ¥
print(iris.data.shape) #dataì— ëŒ€í•š ì „ì²´ í–‰, ì—´ ê¸¸ì´ë¥¼ ì¶œë ¥ (150ê°œì˜ ìë£Œ, 4ê°œì˜ column)
print(iris.feature_names) #feature(column) name ì¶œë ¥
#'sepal length (cm)' : ê½ƒë°›ì¹¨ ê¸¸ì´
#'sepal width (cm)' : ê½ƒë°›ì¹¨ ë„ˆë¹„
#'petal length (cm)' : ê½ƒì ê¸¸ì´
#'petal width (cm)' : ê½ƒì ë„ˆë¹„
```

```python
print(iris.DESCR) #ë°ì´í„°ì— ì„¸ë¶€ì ì¸ ë‚´ìš©ì„ ì €ì¥í•˜ê³  ìˆëŠ” DESCRì„ ì¶œë ¥
#(sklearn datasetì€ í•´ë‹¹ datasetì— ì„¸ë¶€ì •ë³´ë¥¼ ì œê³µ)
```

```python
## ë°ì´í„° í•™ìŠµ ì‹œí‚¤ê¸°
x = iris.data[:, :2] #ì „ì²´ í–‰ê³¼ 2ê°œì˜ ì—´ #ì²˜ìŒ 2ê°œì˜ columnë§Œ ìˆëŠ” ì „ì²´ ë°ì´í„°
y = iris.target #target(ì •ë‹µ)
SVM = svm.SVC(kernel="linear", C=1).fit(x, y) #sklearn.svm.SVC() : SVM ëª¨ë¸ ìƒì„± #í•™ìŠµ
# (1) kernel : ì°¨ì›ì˜ ì„¸ë¶„í™”ë¥¼ í•¨. ëŒ€í‘œì ìœ¼ë¡œ rbf(Radial Basis Function), linear, poly(Polynomial)ì„ ì‚¬ìš©. ê¸°ë³¸ê°’ rbf
# (2) C : ê°’ì„ ë‚®ì¶”ë©´ ì´ˆí‰ë©´ì´ ë§¤ë„ëŸ¬ì›Œì§€ê³ , ê°’ì„ ë†’ì´ë©´ ì„œí¬íŠ¸ ë²¡í„°ë“¤ì„ ë” ì˜ ë¶„ë¥˜í•¨. ê¸°ë³¸ê°’ 1.0
#         Cê°’ì„ ì‘ê²Œ í•˜ë©´ í•œìª½ì˜ ì˜¤ë¶„ë¥˜ë¥¼ ë‚®ì¶”ëŠ” ëŒ€ì‹ ì— ë‹¤ë¥¸ìª½ ì˜¤ë¶„ë¥˜ê°€ ë†’ì•„ì§€ê³ , Cê°’ì„ í¬ê²Œ í•˜ë©´ ì–‘ìª½ì˜ ì˜¤ë¶„ë¥˜ë¥¼ ë¹„ìŠ·í•˜ê²Œ í—ˆìš©
# (3) gamma : ê°’ì„ ë‚®ì¶”ë©´ ì´ˆí‰ë©´ì—ì„œ ë©€ë¦¬ ë–¨ì–´ì§„ ì„œí¬íŠ¸ ë²¡í„°ë“¤ì˜ ì˜í–¥ì´ ë‚®ê³  ê°’ì„ ë†’ì´ë©´ ë©€ë¦¬ ë–¨ì–´ì§„ ìš”ì†Œë“¤ì˜ ê°’ì´ ì˜í–¥ì´ ì»¤ì§. ê¸°ë³¸ê°’ auto
```

```python
## ë°ì´í„° ì‹œê°í™” ì „ì²˜ë¦¬
x_min, x_max = x[:, 0].min()-1, x[:, 0].max()+1 #ì²«ì§¸ columnê°’ ì¤‘ ìµœì†Œê°’ -1, ìµœëŒ€ê°’ +1 #numpy.min(), numpy.max()
y_min, y_max = x[:, 1].min()-1, x[:, 1].max()+1 #ë‘˜ì§¸ columnê°’ ì¤‘ ìµœì†Œê°’ -1, ìµœëŒ€ê°’ +1 #numpy.min(), numpy.max() #íƒ€ê²Ÿì´ ì•„ë‹˜
#plotì˜ ì²˜ìŒê³¼ ëì„ ì§€ì •í•´ ì£¼ê¸° ìœ„í•œ ì „ì²˜ë¦¬

plot_unit = 0.025 #ì¶•ì˜ ë‹¨ìœ„ #x_max ê°’ì„ x_min ê°’ìœ¼ë¡œ ë‚˜ëˆˆ í›„ì— 100ìœ¼ë¡œ ë‹¤ì‹œ ë‚˜ëˆ„ì–´ ë‚˜ì˜¨ ê°’ 0.025
#ì¶•ì˜ ë‹¨ìœ„ëŠ” plotì„ ê·¸ë¦´ ë•Œ ë°°ê²½ì˜ ì„¸ë°€í•œ ì •ë„ë¥¼ ë‚˜íƒ€ë‚´ê¸° ë•Œë¬¸ì— ê·¸ ìˆ˜ì¹˜ë¥¼ ë‚®ê²Œ ì¡ì•„ì„œ ì„¸ë°€í•˜ê²Œ í™•ì¸ì„ í•œë‹¤.
#ë°©ì•ˆì§€ì˜ ëˆˆê¸ˆ í•˜ë‚˜ì˜ í¬ê¸°. (ê°€ë¡œì¶•/ì„¸ë¡œì¶• ê°ê°€ì˜ ì¢Œí‘œ í¬ê¸° : 0.025)
#x_minê³¼ x_max ì‚¬ì´ì— plot_unit(0.025)ë§Œí¼ì˜ ê°„ê²©ì„ ë‘” ìˆ«ìë¥¼ ìƒì„±
#y_minê³¼ y_max ì‚¬ì´ì— plot_unit(0.025)ë§Œí¼ì˜ ê°„ê²©ì„ ë‘” ìˆ«ìë¥¼ ìƒì„±
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_unit), np.arange(y_min, y_max, plot_unit))
#xx : xê°€ yì˜ ì°¨ì› í¬ê¸°ì˜ í–‰ì˜ ê°œìˆ˜ ë§Œí¼ ë°˜ë³µëœ ê°’ì„ ê°€ì§€ê³  ìˆëŠ” ë³€ìˆ˜
#yy : yê°€ xì˜ ì°¨ì› í¬ê¸°ì˜ ì—´ì˜ ê°œìˆ˜ ë§Œí¼ ë°˜ë³µëœ ê°’ì„ ê°€ì§€ê³  ìˆëŠ” ë³€ìˆ˜
#np.meshgrid() : 1ì°¨ì› ë°°ì—´í˜•íƒœì˜ ë°ì´í„°ë“¤ì´ ì£¼ì–´ì§€ë©´ 2ì°¨ì› ë°°ì—´í˜•íƒœë¡œ êµì²´í•´ì£¼ëŠ” í•¨ìˆ˜
#np.arange() : ì§€ì •ëœ ê°„ê²© ë‚´ì—ì„œ ê· ë“±í•˜ê²Œ ê°„ê²©ì„ ë‘” ê°’ì„ ëŒë ¤ì¤Œ #pythonì˜ range()ì™€ ë¹„ìŠ·

#https://datascienceschool.net/view-notebook/17608f897087478bbeac096438c716f6/
```

```python
print("x_min : ", x_min)
print("x_max : ", x_max)
print("y_min : ", y_min)
print("y_max : ", y_max)
print("xx : ", xx) #x_minì—ì„œ x_maxê¹Œì§€ plot_unitì”© ì¦ê°€ (ê°€ë¡œ: ê°€ë¡œì¶•ì´ë¯€ë¡œ) #2ì°¨ì› #ë°ì´í„° ìˆ˜ 150ì—ì„œ ì¢Œìš° ì—¬ë¶„ë§Œí¼ ì¡°ê¸ˆ ë” ì¦ê°€
print("yy : ", yy) #y_minì—ì„œ y_maxê¹Œì§€ plot_unitì”© ì¦ê°€ (ì„¸ë¡œ: ì„¸ë¡œì¶•ì´ë¯€ë¡œ) #2ì°¨ì› #ë°ì´í„° ìˆ˜ 150ì—ì„œ ìƒí•˜ ì—¬ë¶„ë§Œí¼ ì¡°ê¸ˆ ë” ì¦ê°€
```

## 머신러닝

- (1) 특징(feature)
    - 각각의 아이템을 설명하는데 사용하는 구분 가능한 특성 또는 특징 의 개수 (ex. 자동차와 오토바이의 특징 = 바퀴의 개수, 차체의 크기)

- (5) 학습 셋(training set)
    - 학습에 사용하는 데이터 셋

#### 머신러닝 기술
- (1) 분류(classification) : 데이터로부터 클래스를 예측
    - 미리 정의된 카테고리로 데이터 분류 (ex. 일반종양/악성종양, 스팸/비스팸, 찬성/반대)
    - 데이터로는 텍스트, 이미지 사용
    - 나이브 베이즈, 랜덤 포레스트, 서포트 벡터 머신
- (2) 군집(clustering) : 데이터로부터 의미 있는 그룹을 나눔 
    - 서로 더 비슷한 집합을 동일한 그룹으로 그룹화
    - 개체는 사전에 정의되어 있지 않음 (ex. 1반 남자, 1반 여자, 2반 남자, 2반 여자를 각각 남/녀, 1반/2반으로 나눌 수 있다.)
    - K-Means 군집, 계층적 군집(서포트 벡터 머신)
- (3) 회귀(regression) : 데이터 분석을 통해 값을 예측
    - 하나의 변수 평균값과 다른 변수 해당 값간 관계 측정
    - 변수 간의 관계 추정 위한 통계
    - 학습 데이터 이용해 결과값 예측
    - 선형회귀, 로지스틱 회귀
    
#### 데이터 셋의 종류
![week4_6.png](images/week4_6.png)
- Training set : 모델 훈련에 사용하는 데이터 셋.
- Validation set : 모델 훈련에 적절한 지점을 찾기 위해 사용하는 데이터 셋으로 과적합(over fitting) 또는 과소적합(under fitting) 방지하기 위한 stopping point를 찾음
- Test set : 모델의 성능을 평가하기 위해 사용하는 데이터 셋

#### 데이터 셋 구성 방법
- 1. Holdout : 데이터 셋의 일부를 훈련용 데이터, 나머지 일부를 테스트용 데이터로 구성
    ![week4_7.png](images/week4_7.png)
    - 데이터 셋을 train set, test set 두 셋으로 나눔
    - train set과 test set의 일정 비율을 설정
    - train set이 작으면 모델 정확도의 분산이 커짐
    - train set이 커지면 test set으로부터 측정한 정확도의 신뢰도 하락
    - 위와 같은 단점을 극복하기 위해 holdout을 반복적으로 실행하는 것이 random sampling
- 2. Random subsampling : Holdout 방식을 반복
    ![week4_8.png](images/week4_8.png)
    - k개의 부분 데이터 셋을 사용함
        - 각 부분 데이터 셋은 랜덤으로 정해짐
        - 각 실험마다 training set으로 모델을 훈련, test set으로 모델을 테스트
    - 최종 성능은 각 실험 성능의 평균으로 도출
- 3. K-Cross validation : 중복되지 않는 k개의 부분 셋으로 나누어 구성
    ![week4_9.png](images/week4_9.png)
    - data set을 k개의 fold로 나누어 사용
        - k번의 실험을 진행하며, k-1개 fold는 training set, 1개 fold는 test set으 로 사용
    - random subsampling과 유사하지만, k-fold cross validation을 사용할 경우 모든 데이터를 train과 test에 적용할 수 있음
    - 최종 성능은 k번의 실험 성능의 평균으로 도출
- 4. Stratified sampling : 각 클래스로부터 일정 비율의 샘플 추출
    - stratified sampling, 층별 표집 방법
    - 데이터를 클래스에 따라 각 그룹으로 분리 후, 각 그룹으로부터 일정 비율의 샘플을 무작위로 추출하는 방법
    - 전체 데이터에서 무작위로 샘플을 추출할 경우 표본이 특정 클래스에 편중될 수 있는 단점이 있음
    - stratified sampling 방법을 통해 위와 같은 단점을 보완하며 샘플의 대표성을 높임
- 5. Bootstrap : 중복을 허용하는 샘플 추출
    - bootstrap 방법은 샘플 추출 시 각 데이터의 중복 추출을 허용함
    - 전체 데이터에서 n개의 샘플을 추출하는데, 다음 시도에서 기존에 추출 했던 샘플이 다시 추출될 수 있음
    - 중복을 허용한 샘플 추출이 반복적으로 시행됨
    - 최종 성능은 실행된 실험 성능들의 평균으로 도출됨
    
#### 모델 성능평가 척도
![week4_10.png](images/week4_10.png)
- Confusion matrix: 모델의 성능을 통계적인 수치로 시각화 하는 방법 (위 표가 Confusion matrix)
- Accuracy(정확도) : 모델이 정확하게 분류 또는 예측하는 데이터의 비율 
    - ${a + d \over a + b + c + d} = {TP + TN \over TP + TN + FP + FN}$
    - 10000건의 데이터 중 class yes가 9900건, class no가 100건 인 경우, 모두 yes로 예측해도 정확도는 99%가 나오므로 정확도는 한계가 있다.
- Precision(정확률, 저항률, 정밀도) : 모델이 검출한 데이터 중 올바르게 검출된 데이터의 비율 
    - ${a \over a + c} = {TP \over TP + FP}$
    - precision과 recall, 두 수치가 균형을 유지하는 모델이 우수한 모델이다.
- Recall(재현율) : 실제 해당 데이터 중 모델이 올바르게 검출한 데이터의 비율 
    - ${a \over a + b} = {TP \over TP + TN}$
    - precision과 recall, 두 수치가 균형을 유지하는 모델이 우수한 모델이다.
- F-Measure : precision과 recall을 통합해 나타내는 정확도 
    - ${2a \over 2a + b + c} = {2TP \over 2TP + TN + FP}$
- Learning Curve : 모델의 성능 시각화
    - learning curve(학습 곡선)는 모델의 성능을 시각화해 확인가능
    - 데이터 셋의 크기에 따른 오류의 변화를 확인
    - Train set의 크기가 매우 작을 경우 오류는 매우 적겠지만, 크기가 커질수록 오류가 증가할 것
    - 반면, cross validation의 경우 적은 데이터 셋으로는 오류가 많이 발생하는 반면, 샘플의 크기가 커질수록 일반화가 되면서 오류가 감소할 것
    - 과소적합의 경우 learning curve
    ![week4_11.png](images/week4_11.png)    
        - holdout 방법을 사용할 경우 train set의 크기가 커질수록 오류가 증가하 지만, 어느 수치 이상으로 증가하지 않음
        - cross validation 방법을 사용할 경우 train set의 크기가 커질수록 오류가 감소하지만, 어느 수치 이하로는 감소하지 않음
    - 과적합의 경우 learning curve
    ![week4_12.png](images/week4_12.png)
        - holdout과 cross validation 방법 간의 오차의 차이(gap)이 쉽게 좁혀지지 않음
        - 많은 데이터 셋을 사용할 때 차이가 좁혀짐
        - 과적합 모델의 경우 더 많은 데이터를 수집해 사용하는 것이 필요
- Receiver Operating Characteristic(ROC) Curve : 민감도와 특이도의 관계를 파악할 수 있는 그래프 
    - ROC curve의 y축에 TP, x축에 FP 수치를 입력해 두 수치의 균형을 살펴 볼 수 있음
    - 각 모델의 성능은 roc curve 상의 한 점으로 표현
    - Y축 수치 : TP
        - Sensitivity(민감도), True Positive Rate (TPR)
        - 실제 yes 클래스 데이터 중 모델이 예측한 yes 클래스 데이터의 비율 $TPR = {TP \over P}$
        - Sensitivity = Recall
    - X축 수치 : FP
        - Specificity(특이도), True Negative Rate (TNR)
        - 실제 yes 클래스 데이터 중 모델이 예측한 no 클래스 데이터의 비율
        - Specificity = 1 - Sensitivity
    - 1 – Specificity
        - False Positive Rate (FPR)
        - 실제 no 클래스 데이터 중 모델이 yes 클래스로 예측한 데이터의 비율 $FPR = {FP \over N}$
    ![week4_13.png](images/week4_13.png)
    모델이 yes 클래스를 정확하게 예측할수록 sensitivity, TPR 값이 높아짐    
    roc curve 아래 면적인 area under curve (auc)가 1에 가까워질수록 모델이 yes 클래스를 예측하는 정확도가 높음
- Residuals
    - 회귀분석 모델의 예측 값과 실제 값의 차이 즉, 오차(error)를 잔차(residual)라고 함
    - Residual plot
        - residual plot은 세로축에 residual, 가로축에 독립변수를 나타내는 산포도
        - 산포도를 통해 어떤 선형 모델이 데이터에 적합한지 알 수 있음
    - 적합한 선형 모델 : residual이 x축을 기준으로 랜덤으로 분포하는 상태
    - 적합한 비선형 모델 : residual이 패턴을 보이며 분포하는 상태
    ![week4_14.png](images/week4_14.png)
- Root Mean Squared Error(RMSE) : 모델 예측 값의 오차 계산
    - MSE, 평균 제곱 오차
    - MSE는 회귀선과 모델 예측 값 사이의 오차(잔차, residual)를 사용 
    - 오차를 제곱한 값들의 평균
    - $MSE = {\sum_{i=1}^n (\hat{y}-y_i)^2 \over n}$
    - MSE에 루트를 씌운것이 RMSE

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
```

```python
#데이터 셋 분리 방법 관련 모듈

from sklearn.model_selection import train_test_split #training set과 test set으로 분리하기 위한 train test split
from sklearn.model_selection import StratifiedKFold #stratified k fold cross validation을 사용하기 위한 stratified k fold(데이터 클래스 비율 유지)
#일반 Kfold 모듈과는 다르게 각 fold내 데이터의 클래스 비율을 일정하게 유지한다.
from sklearn.model_selection import cross_val_score #cross validation 결과의 정확도 측정 위한 cross val score
```

```python
# 성능 평가 관련 모듈

from sklearn.metrics import confusion_matrix #분석 결과의 confusion matrix를 추출하기 위한 모듈
from sklearn.metrics import accuracy_score #분석 결과의 accuracy를 측정하기 위한 모듈
from sklearn.metrics import classification_report #분석 결과의 recall, precision, f-measure를 측정하기 위한 모듈
from sklearn.metrics import roc_auc_score #roc 곡선 아래 넓이를 구하기 위한 모듈
from sklearn.metrics import mean_squared_error #분석 결과의 MSE를 구하기 위한 모듈
```

```python
data = datasets.load_breast_cancer() #유방암 데이터 로드
X = data.data #속성 데이터(숫자)
y = data.target #클래스 데이터(0/1, 악성-양성)
```

### holdout

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) #train:test = 8:2
```

```python
#decision tree 분류 모델 생성
clf = DecisionTreeClassifier() #모델 생성
clf.fit(X_train, y_train) #학습 데이터로 학습
clf #모델 학습 완료
```

```python
y_pred = clf.predict(X_test) #학습된 모델에 테스트 셋으로 예측
```

```python
#모델 성능 평가
print("Confusion Matrix")
print(confusion_matrix(y_test, y_pred)) #결과를 confusion_matrix로 나타낸다. 정답이 y_test, 모델 예측값이 y_pred
```

```python
print("Accuracy")
print(accuracy_score(y_test, y_pred, normalize=True)) 
#normalize=True하면 올바르게 분류된 데이터 비율 출력
#normalize=False하면 올바르게 분류된 데이터의 건수 출력
```

```python
print("Classification Report")
print(classification_report(y_test, y_pred))
#f1-score가 f-measure
```

```python
print("AUC")
print(roc_auc_score(y_test, y_pred)) #roc곡선의 아래 면적 출력
```

```python
print("Mean Squared Error")
print(mean_squared_error(y_test, y_pred))
```

### k fold cross validation

```python
skf = StratifiedKFold(n_splits=10) #stratified k fold 생성. 10개의 folds.
skf.get_n_splits(X, y) #stratified k fold에 데이터를 넣어 fold를 분리한다.
print(skf)
```

```python
for train_index, test_index in skf.split(X, y): #데이터 확인
    print("train set: ", train_index)
    print("test set: ", test_index)
```

```python
clf = DecisionTreeClassifier()
scores = cross_val_score(clf, X, y, cv=skf)
print("K Fold Cross Validataion Score")
print(scores) #각 점수들
print("Average Accuracy")
print(scores.mean()) #점수들의 평균
```

### k fold cross validation - shuffle

```python
skf_sh = StratifiedKFold(n_splits=10, shuffle=True) #stratified k fold 생성. 10개의 folds. shuffle 추가
skf_sh.get_n_splits(X, y) #stratified k fold에 데이터를 넣어 fold를 분리한다.
print(skf_sh)  
```

```python
for train_index, test_index in skf_sh.split(X, y): #데이터 확인
    print("train set: ", train_index)
    print("test set: ", test_index)
```

```python
clf = DecisionTreeClassifier()
scores = cross_val_score(clf, X, y, cv=skf_sh)
print("K Fold Cross Validation Score")
print(scores) #각 점수들
print("Average Accuracy")

print(scores.mean()) #평균
```

```python
from sklearn import linear_model
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# matplotlib에는 사전에 정의한 많은 스타일이 있다.
# ggplot도 그 중 하나임
# R에서 사용하는 ggplot을 흉내내는 스타일을 사용하겠다고 선언

matplotlib.style.use('ggplot')
```

```python
data = {"x": [59, 49, 75, 54, 78, 56, 60, 82, 69, 83, 88, 94, 47, 65, 89, 70],
        "y": [209, 180, 195, 192, 215, 197, 208, 189, 213, 201, 214, 212, 205, 186, 200, 204]}

df = pd.DataFrame(data)
df.head()
```

```python
df.plot(kind="scatter", #산점도
        x="x", #x는 df의 x컬럼
        y="y", #y는 df의 y컬럼
        figsize=(5, 5), #가로 세로 5인치
        color="blue") #점 색상
```

```python
lr = linear_model.LinearRegression() #Linearregression()
lr.fit(pd.DataFrame(df["x"]), pd.DataFrame(df["y"])) #독립변수 x, 종속변수y

# 선형 회귀식의 세로축 절편 'intercept_'를 구하여 출력한다.
print("a value = ", lr.intercept_)

# 선형 회귀식의 기울기 'coef_'를 구하여 출력한다.
print("b balue = ", lr.coef_)
```

```python
prediction = lr.predict(pd.DataFrame(df["x"])) #위의 모델로 값 예측

residuals = pd.DataFrame(data["y"]) - prediction
print(residuals) #실제 y값과 예측한 y값을 비교하여 잔차(residuals)를 구한다.

residuals.describe() # 변수의 갯수(6개), 잔차의 평균값, 잔차의 표준편차, 최소값, 25% 값, 50% 값, 75% 값, 최대값을 출력한다.
```

```python
#잔차를 제곱하여 전체를 합침. 결과값을 SSE라는 변수에 할당
SSE = (residuals**2).sum()
print("SSE = ", SSE)

#y값의 표준편차를 제곱한 것을 모두 합침. 그 결과값을 SST라는 변수에 할당
SST = ((df["y"]-df["y"].mean())**2).sum()
print("SST = ", SST)

#결정계수 R을 구함
R_squared = 1 - (SSE/SST)
print('R_squared = ', R_squared)
```

```python
df.plot(kind="scatter", x="x", y="y", figsize=(5,5), color="red")

# Plot regression line
plt.plot(df["x"], prediction, color="blue")
```

```python
#sklearn.metrics라는 패키지로부터 mean_squared_error 모듈을 불러들임
from sklearn.metrics import mean_squared_error

#결정계수 R값을 구함
print('score = ', lr.score(X = pd.DataFrame(data["x"]), y = data["y"]))

#실제값(data[y])과 회귀식 값(prediction)의 차이의 제곱을 구함
print('Mean_Squared_Error = ', mean_squared_error(prediction, data['y']))

#Mean squared error의 제곱근 값을 구함
print('RMSE = ', mean_squared_error(prediction, data['y'])**0.5)
```

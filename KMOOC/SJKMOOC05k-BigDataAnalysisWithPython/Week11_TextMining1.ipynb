## Text Mining

#### 텍스트 정규화(Text Normalization)
- 텍스트의 형태를 일관되게 변형하는 작업
- 토큰화(Tokenization)
    - 텍스트를 의미 단위(토큰)로 분할하는 작업
- 어간 추출(Stemming)
    - 형태가 변형된 단어로부터 어간을 분리하는 작업

#### 형태소 분석(POS-Tagging)
- 토큰의 형태소를 파악하는 작업

#### 텍스트 분석의 종류
- 정보 추출(Information Retrieval) 
    - 문서들을 특정 분류 체계에 따라 분류하는 작업
- 감성 분석 (Sentiment Analysis)
    - 문서내의 정형 데이터를 추출하는 작업
- 정보 추출 (Information Retrieval)
    - 문서에 내포되어 있는 감정과 의견을 추출하는 작업 
    
#### 토큰화
- 텍스트를 의미 단위로 분할하는 작업으로, 단위를 토큰(token)이라고 함
- Type : 단어 요소, 어간
- Token : 텍스트 내부의 원소
- 예 : Text normalization is the process of transforming text into a single canonical form that it might not have had before 
    - Token은 공백을 기준으로 단순 분할
    - type은 중복을 제외하고 어간이 같은 단어의 중복을 제외(have - had)
- 영어 : 단순 공백 기준 분할
- 한글 : 품사 기준으로 토큰화
    - 예 : 한국어를 토큰화하는 예시 문장 입니다.
    - 토큰화 : 한국어, 를, 토큰화, 하다, 예시, 문장, 입, 니다

#### Lemmatization
- 문장 속에서 다양한 형태로 변형된 단어의 표제어(lemma)를 찾는 작업
- 표제어 : 단어의 의미를 사전에서 찾을 때 사용하는 기본형 (Stemming -> stem)
- 어간 추출(stemming)과의 차이점 
    - 어간 추출 작업은 단어의 단순한 어근을 추출하는 작업
    - Lemmatization은 문장 구조상 단어의 의미를 이해한 후 기본형 추출 
    - Stemming 예: flies → fly 
    - Lemmatization 예: flies가 "날다"(동사)인지 "파리"(명사)인지 파악 
    
#### Stemming
- 어간 추출(Stemming)
    - 형태가 변형된 단어에 대해 접사 등을 제거하고 어간을 분리해 내는 작업 
    - 예시 : fly, flying, flies -> fly
    - stemming, stemmed, stemmer -> stem
- 포터 알고리즘(Porter‟s Stemming Algorithm) 
    - 대표적인 영어 어근 추출 알고리즘
    - 1a. 복수형은 단수형으로
        - sses → ss : caresses → caress 
        - ies → i :  ponies → poni 
        - ss → ss : caress → caress 
        - s → NULL : cats → cat 
    - 1b. 동사를 원형으로
        - eed → ee : agreed → agree
        - ed → NULL : plastered → plaster / bled → bled 
        - ing → NULL : motoring → motor / sing → sing 
    - 1b1. 단어를 기본형으로
        - at → ate : conflat(ed) → conflate 
        - bl → ble : troubl(ing) → trouble 
        - iz → ize : siz(ed) → size 
        - … 
    - 2. 기본형으로 바꿔준다.
        - ational → ate : relational → relate 
        - tional → tion : conditinal → condition 
        - … 
        
#### 형태소 분석
- 형태소 분석
    - POS-Tagging (Part-Of-Speech Tagging)
    - 토큰의 품사 정보를 추출하는 작업 
- Open class vs. Closed class 
    - Open class : 새로운 단어의 등장이 상대적으로 많은 품사
        - 영어의 open class: Noun, Verbs, Adjectives, Adverbs 
        - 명사: Sejong University, Samsung, cat, table, … 
        - 동사: listen, see, enter, … ○형용사: good, better, worse, … 
        - 부사: slowly, hardly, … 
    - Closed class : 상대적으로 고정되어 있는 셋
        - 전치사: of, in, at, … 
        - 조동사: may, have, can, will, must, … 
        - 대명사: I, you, she, mine, their, … 
        - 주로 문장 내에서 문법적인 역할을 하는 단어 
- 영어 형태소 분석기
    - tanford POS-Tagger, NLTK POS-Tagger 
- 한국어 형태소 분석기 
    - Hannanum, Kkma, Komoran, Twitter 
    
#### 정보 추출
- 정보 추출 (Information Retrieval, IR) 
    - 구조화 되지 않은 문서 군집 내에서 필요한 정보를 포함하고 있는 문서를 찾아내는 작업 
    - 예: web search 
    - 각 단어는 벡터로 표현 가능
- Bag of words
    - 문서 내에 등장하는 단어들의 순서를 무시
    - 예 : John is quicker than Mary. Mary is quicker than John. 
        - 위 두 문장은 같은 벡터를 가지고 있다.
        - (John, Mary, is, quicker, than) 
        - 등장하는 단어의 순서는 고려하지 않음 
- Term Frequency (TF)         
    - 특정 단어가 특정 문서 내에 등장하는 빈도
    - 값이 클수록 문서에 많이 등장하는 단어이지만, 정보가 부족한 단어일 가능성이 높다.
    - 따라서 문서를 구분하는 척도로 사용하기에 적절하지 못하다.
    - 드물게 등장하는 단어가 더 중요한 정보를 가지고 있다.
    - 드문 단어들에 대해 가중치를 부여할 필요가 있다.
- Inverse Document Frequency (IDF) 
    ![week11_1.png](images/week11_1.png)
    - 각 문서를 tf-idf 값으로 이루어진 벡터로 표현할 수 있다.
    - 문서는 벡터 공간에서 점 또는 벡터로 표현된다.

- 새로운 단어가 등장하더라도 해당 빆도에 1을 더해 줌으로써 확률이 0이 되는 것을 방지

## 전처리 

```python
import nltk #영문 자연어 처리 패키지
```

```python
print(df.ix[0]["description"]) #한 개의 뉴스 데이터를 사용
#0번 인덱스 행의 description
```

```python
tokens = nltk.word_tokenize(df.ix[0]["description"]) #단순 공백 기반 토큰화
tokens
```

```python
tokens = [token.lower() for token in tokens if len(token) > 1] #각 토큰마다 길이가 1보다 긴 토큰들을 소문자로 변경해 저장
tokens
```

```python
#bi-gram : 단어들을 두 개씩 묶어 하나의 토큰으로 추출한 것 
tokens_bigram = nltk.bigrams(tokens)
for token in tokens_bigram :
    print(token)
```

```python
#tri-gram : 단어들을 세 개씩 묶어 하나의 토큰으로 추출한 것 
tokens_trigram = nltk.trigrams(tokens)
for token in tokens_trigram :
    print(token)
```

## 특징값 추출

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#TfidfVectorizer 모듈을 사용해 텍스트 데이터의 TF-IDF 값으로 문서 단어 행렧을 구성 
```

```python
text_data_list = df['description'].astype(str).tolist()
# astype 함수를 사용해 문자열로, tolist 함수를 사용해 리스트로 변환 후 text_data_list 변수에 저장 
text_data_arr = np.array([''.join(text) for text in text_data_list])
```

```python
vectorizer = TfidfVectorizer(min_df=2, ngram_range=(1, 2), strip_accents='unicode', norm='l2')
#TfidfVectorizer함수를 사용해 tf-idf 문서 단어 행렬을 맊드는 모듈 vectorizer 선얶 
#Min_df=2 : 단어 최소 등장 빆도로 2번 이상 등장하는 단어들을 대상으로 함 
#Ngram_range : 단어 추출 단위로 (1,2)는 unigram과 bigram 추출, 1<= n <= 2 
#Strip_accents=„Unicode‟ : accents(억양표시)를 제거하며 unicode에 해당하는 모든 문자에 적용 가능 
text_data = vectorizer.fit_transform(text_data_arr)
#vectorizer의 Fit_transform 함수를 사용해 배열에 저장된 데이터의 문서단어행렬을 구하고 
#matrix 형식 변수 text_data에 저장
```

```python
df_tfidf = pd.DataFrame(text_data.A, columns=vectorizer.get_feature_names())
df_tfidf
```

## 뉴스 분류


```python
#성능 측정 패키지 임포트
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
#데이터 셋 준비
description = df['description'].astype(str).tolist() #내용
category = df['category'].astype(str).tolist() #카테고리
```

```python
trainset_size = int(round(len(description)*0.80))

x_train = np.array([''.join(data) for data in description[0:trainset_size]])
y_train = np.array([data for data in category[0:trainset_size]])

x_test = np.array([''.join(data) for data in description[trainset_size+1:len(description)]])
y_test = np.array([data for data in category[trainset_size+1:len(category)]])
```

```python
X_train = vectorizer.fit_transform(x_train) #training set을 기반으로 문서단어행렬 구성 
X_test = vectorizer.transform(x_test) #구성한 단어들을 기반으로 문서단어행렬 구성
```

```python
#각 분류 모델의 성능을 기록할 데이터 프레임
df_per = pd.DataFrame(columns=['Classifier', 'F-Measure', 'Accuracy'])
df_per
```

## Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB

nb_classifier = MultinomialNB().fit(X_train, y_train)
nb_pred = nb_classifier.predict(X_test)
```

```python
print('\n Confusion Matrix \n')
print(confusion_matrix(y_test, nb_pred))
print('\n Classification Report \n')
print(classification_report(y_test, nb_pred))
```

```python
fm = round(f1_score(y_test, nb_pred, average='weighted'), 2)
ac = round(accuracy_score(y_test, nb_pred, normalize=True), 2)
df_per.loc[len(df_per)] = ['Naive Bayes', fm, ac]
df_per
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt_classifier = DecisionTreeClassifier().fit(X_train, y_train)
dt_pred = dt_classifier.predict(X_test)
```

```python
print('\n Confusion Matrix \n')
print(confusion_matrix(y_test, dt_pred))
print('\n Classification Report \n')
print(classification_report(y_test, dt_pred))
```

```python
fm = round(f1_score(y_test, dt_pred, average='weighted'), 2)
ac = round(accuracy_score(y_test, dt_pred, normalize=True), 2)
df_per.loc[len(df_per)] = ['Decison Tree', fm, ac]
df_per
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(n_estimators=100)
rf_classifier.fit(X_train, y_train)
rf_pred = rf_classifier.predict(X_test)
```

```python
print('\n Confusion Matrix \n')
print(confusion_matrix(y_test, rf_pred))
print('\n Classification Report \n')
print(classification_report(y_test, rf_pred))
```

```python
fm = round(f1_score(y_test, rf_pred, average='weighted'), 2)
ac = round(accuracy_score(y_test, rf_pred, normalize=True), 2)
df_per.loc[len(df_per)] = ['Random Forest', fm, ac]
df_per
```

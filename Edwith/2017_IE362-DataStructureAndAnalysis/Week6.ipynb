### Characteristics of Tree

Num of edge = Num of nodes - 1 (전체 레퍼런스 수 : 전체 노드에서 root를 빼주면 전체 엣지의 수가 나온다.)    
Depth of root = 0 (root의 depth는 0)    
Height of root = height of tree (루트의 height는 트리의 height)     
(Maximum num. of nodes at level i with degree d) = $d^i$ (i번 째 레벨의 최대 노드 수 = degree의 i승)    
(Maximum num. of leaves with height h and degree d) = $d^h$ (i번 째 레벨의 최대 리프 수)    
(Maximum size of a tree with height h and degree d) = $1 + d + d^2 + \ldots + d^h = {d^{h+1} - 1 \over d - 1}$(트리의 최대 사이즈)    
(Height of a complete tree with size s and degree d) = $[\log_d(s(d-1)+1)]-1$

```python
class TreeNode:
    nodeLHS = "" #항상 자신보다 작은 value의 노드를 레퍼런스
    nodeRHS = "" #항상 자신보다 큰 value의 노드를 레퍼런스
    nodeParent = "" #부모 노드를 레퍼런스
    value = ""
    
    def __init__(self, value, nodeParent):
        self.value = value
        self.nodeParent = nodeParent
    def getLHS(self):
        return self.nodeLHS
    def getRHS(self):
        return self.nodeRHS
    def getValue(self):
        return self.value
    def getParent(self):
        return self.nodeParent
    def setLHS(self, LHS):
        self.nodeLHS = LHS
    def setRHS(self, RHS):
        self.nodeRHS = RHS
    def setValue(self, value):
        self.value = value
    def setParent(self, nodeParent):
        self.nodeParent = nodeParent
```

```python
#좌측에 자신보다 작은 값, 우측에 자신보다 큰 값.
class BinarySearchTree:
    root = "" #트리는 root에 대해서만 레퍼런스를 저장하고 있다. 나머지는 루트에서 traverse해 나가면서 찾아낸다.
    
    def __init__(self):
        pass
    
    def insert(self, value, node = ""): #좌, 우측이 비어 있지 않아 재귀될 경우 노드에 값이 들어간다.
        #일반적인 첫 번째 insert때는 항상 node가 Garbage
        if node == "":  #노드가 제대로 정의되어 있지 않다면 #처음으로 시작하는 것이라면
            node = self.root #루트를 받아온다.
        if self.root = "": #루트가 비어 있으면
            self.root = TreeNode(value, "") #현재 노드를 루트로 만든 후
            return #리턴
        if value == node.getValue(): #insert하려는 값이 현재 노드의 값과 같다면
            return #중복된 insert하지 않고 리턴
        if value > node.getValue(): #insert하려는 값이 현재 노드의 값보다 크다면 -> 우측에 저장해야 한다.
            if node.getRHS == "": #우측이 비어 있다면
                node.setRHS(TreeNode(value, node)) #우측 노드에 바로 저장
            else: #우측이 비어 있지 않다면
                self.insert(value, node.getRHS()) #우측 트리에 대한 삽입 재귀
        if value < node.getValue(): #insert하려는 값이 현재 노드의 값보다 작다면 -> 좌측에 저장해야 한다.
            if node.getLHS() == "": #좌측이 비어 있다면
                node.setLHS(TreeNode(value, node)) #좌측 노드에 바로 저장
            else: #좌측이 비어 있지 않다면
                self.insert(value, node.getLHS()) #좌측 트리에 대한 삽입 재귀
        return
    
    def search(self, value, node = ""): #트리는 최대 Max depth만큼 검색을 하면 값을 찾을 수 있다.
        if node == "": #노드가 제대로 정의되어 있지 않다면 #처음으로 시작하는 것이라면
            node = self.root #루트를 받아 온다.
        if value == node.gatValue(): #찾으려는 값이 현재 노드의 값과 같으면
            return True #리턴
        if value > node.getValue(): #찾으려는 값이 현재 노드의 값보다 크다면 -> 우측 검색
            if node.getRHS() == "": #우측 노드의 값이 정의되어 있지 않다면
                return False #False 리턴. (더 이상 검색할 값이 없다.)
            else: #우측이 비어 있지 않다면
                return self.search(value, node.getRHS()) #우측 트리에 대한 감색 재귀
        if value < node.getValue(): #찾으려는 값이 현재 노드의 값보다 작다면 -> 좌측 검색
            if node.getLHS() == "": #좌측 노드의 값이 정의되어 있지 않다면
                return False #False 리턴. (더 이상 검색할 값이 없다.)
            else: #좌측이 비어 있지 않다면
                return self.search(value, node.getLHS()) #좌측 트리에 대한 검색 재귀
            
    def delete(self, value, node = ""):
        #delete에는 3가지의 경우가 있다 BTS에서
        #case 1 : 자식없는 노드 삭제 : 링크드 리스트의 삭제와 동일 단순히 레퍼런스를 끊어주기만 하면 된다.
        #case 2 : 자식이 하나인 노드 삭제 : 레퍼런스를 남아 있는 자식과 연결
        #case 3 : 자식이 둘인 노드 삭제 : RHS의 Min 혹은 LHS의 Max 이용
        if node == "": #노드가 제대로 정의되어 있지 않다면 #처음으로 시작하는 것이라면
            node = self.root #루트를 받아 온다.
        if node.getValue() < value: #지우려는 값이 현재 노드의 값보다 크다면 -> 우측
            return self.delete(value, node.getRHS()) #우측 트리에 대한 삭제 재귀
        if node.getValue() > value: #지우려는 값이 현재 노드의 값보다 작다면 -> 좌측
            return self.delete(value, node.getLHS()) #좌측 트리에 대한 삭제 재귀
        if node.getValue() == value:
            if node.getLHS() != "" and node.getRHS() != "": #노드의 양쪽이 둘다 있을 경우 (case3)
                nodeMin = self.findMin(node.getRHS()) #우측의 최소값을 찾는다.
                node.setValue(nodeMin.getValue())
                self.delete(nodeMin.getValue(), node.getRHS()) #최소값으로 재귀
                return
            parent = node.getParent()
            if node.getLHS() != "": #(case 2) 한쪽(우측)만 없다.
                if node == self.root:
                    self.root = node.getLHS()
                elif parent.getLHS() == node:
                    parent.setLHS(node.getLHS())
                    node.getLHS().setParent(parent)
                else:
                    parent.setRHS(nide.getLHS())
                    node.getLHS().setParent(parent)
                return
            if node.getRHS() != "": #(case 2) 한쪽(좌측)만 없다.
                if node == self.root:
                    self.root = node.getRHS()
                elif parent.getLHS() == node:
                    parent.setLHS(node.getRHS())
                    node.getRHS().setParent(parent)
                else:
                    parent.getRHS(nide.getRHS())
                    node.getRHS().setParent(parent)
                return
            if node == self.root: #(case1)
                self.root = ""
            elif parent.getLHS() == node:
                parent.setLHS("")
            else:
                parent.setRHS("")
            return
        
    def findMax(self, node = ""): #RHS로만 찾아가면 Max를 찾을 수 있다.
        if node == "":
            node = self.root
        if node.getRHS() == "":
            return node
        return self.findMax(node.getRHS())
    
    def findMin(self, node = ""): #LHS로만 찾아가면 Min을 찾을 수 있다.
        if node == "":
            node = self.root
        if node.getLHS() == "":
            return node
        return self.findMin(node.getLHS())
    
    def traverseLevelOrder(self):
        ret = []
        Q = Queue()
        Q.enqueue(self.root)
        while Q.isEmpty() == False:
            node = Q.dequeue()
            if node == "":
                continue
            ret.append(node.getValue())
            if node.getLHS() != "":
                Q.enqueue(node.getLHS())
            if node.getRHS() != "":
                Q.enqueue(node.getRHS())
        return ret
    
    def traverseInOrder(self, node = ""):
        if node == "":
            node = self.root
        ret = []
        if node.getLHS() != "":
            ret = ret + self.traverseInOrder(node.getLHS()) #재귀로 좌측부터 탐색
        ret.append(node.getValue()) #Current node
        if node.getRHS() != "": #좌측 이후, 재귀로 우측으로 탐색
            ret = ret + self.traverseInOrder(node.getRHS())
        return ret
    
    def traversePreOrder(self, node = ""):
        if node == "":
            node = self.root
        ret = []
        ret.append(node.getValue()) #Current node
        if node.getLHS() != "":
            ret = ret + self.traverseInOrder(node.getLHS()) #재귀로 좌측부터 탐색
        if node.getRHS() != "": #좌측 이후, 재귀로 우측으로 탐색
            ret = ret + self.traverseInOrder(node.getRHS())
        return ret
    
    def traversePostOrder(self, node = ""):
        if node == "":
            node = self.root
        ret = []
        if node.getLHS() != "":
            ret = ret + self.traverseInOrder(node.getLHS()) #재귀로 좌측부터 탐색
        if node.getRHS() != "": #좌측 이후, 재귀로 우측으로 탐색
            ret = ret + self.traverseInOrder(node.getRHS())
        ret.append(node.getValue()) #Current node
        return ret
```
